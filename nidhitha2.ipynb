**2. load** **the** **dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
df
```

**3**.**Perfrom** **Data** **Visualization**

1. Scatter plot

```python
sns.scatterplot(x="alcohol", y="speeding", data=df)
```

Inference: In the above graph Speeding increases as alcohol increses. Alcohol is taken on x-axis and speeding is taken on y-axis.

2. Line Plot

```python
sns.lineplot(x="alcohol",y="not_distracted",data=df)
```

Inference: Line plot with alcohol on x-axis and not_distracted on y-axis is plotted. In the graph not_distracted increases as alcohol increases.

3. Distribution Plot

```python
sns.distplot(df["no_previous"])
```

Inference: In this graph x-axis represents no_previous car crashes and y-axis represents the probability density. The density of previous car crashes increases until 15 and later decreases.

4. Relational plot

```python
sns.relplot(x="total", y="ins_premium", data=df, hue="ins_losses")
```

Inference: Relational plot with total on x-axis and ins_premium on y-axis with hue as ins_losses is plotted. It allows us to visualise how variables on a dataset relate to each other

5. Bar Plot

```python
sns.barplot(data=df, x="not_distracted",y="ins_losses",errorbar=None)
sns.figure(figsize(17,8))
```

Inference: Bar graph is plotted with not_distracted on x-axis and ins_losses on y-axis.

6. Joint Plot

```python
sns.jointplot(x="speeding", y="ins_premium", data=df)
```

Inference: Joint plot with speeding on x axis and ins_premium on y-axis is plotted.

7. Box Plot

```python
sns.boxplot(x="speeding",y="alcohol",data=df)
```

Inference: Box plot with speeding on x axis and alcohol on y-axis is plotted.

**Correlation**

```python
cr=df.corr()
cr
```

8. Heat map

```python
sns.heatmap(cr, annot=True,cmap="YlGnBu")
```

Inference: Values above 0.5 are highly correlated and below 0.5 are less correlated.

9. Scatter plot

```python
sns.scatterplot(x="total",y="ins_losses",data=df)
```

Inference: Scatter plot with total on x-axis and ins_losses on y-axis is plotted. In the graph ins_losses is maximum when total is 15.0 and 20.0.

10. Line plot

```python
sns.lineplot(x="no_previous", y="total", data=df)
```

Inference: In this graph, as no_previous car crashes increases total also increases.
